```python
import requests
import json

def request_extractor(img_id_list):
    list_keys= []
    for img_id in img_id_list:
        list_keys.append({
                "key": img_id,
                "resourceId": img_id,
                "expiresIn": "86400"
            }
        )
    
    json_data = {
        "action": "get",
        "keys": list_keys,
        "bucket": "rz-dre-weights",
        "provider": "aws"
    }

    headers = {'Content-Type': 'application/json', 'rz-cuid': 'roadzen-via'}

    response = requests.post('https://baadal.roadzen.xyz/api/v1/files/signedurls',
                  data =json.dumps(json_data), headers=headers).json()
    return response
```
